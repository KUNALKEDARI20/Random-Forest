**About the data:**  
Let’s consider a Company dataset with around 10 variables and 400 records.
The attributes are as follows:   
 Sales -- Unit sales (in thousands) at each location  
 Competitor Price -- Price charged by competitor at each location  
 Income -- Community income level (in thousands of dollars)  
 Advertising -- Local advertising budget for company at each location (in thousands of dollars)  
 Population -- Population size in region (in thousands)  
 Price -- Price company charges for car seats at each site  
 Shelf Location at stores -- A factor with levels Bad, Good and Medium indicating the quality of the shelving location for the car seats at each site  
 Age -- Average age of the local population  
 Education -- Education level at each location  
 Urban -- A factor with levels No and Yes to indicate whether the store is in an urban or rural location  
 US -- A factor with levels No and Yes to indicate whether the store is in the US or not

**Problem Statement:**  
A cloth manufacturing company is interested to know about the segment or attributes causes high sale.
Approach - A Random Forest can be built with target variable Sales (we will first convert it in categorical variable) & all other variable will be independent in the analysis.  


```python
#Import Libraries
import pandas as pd
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection  import cross_val_score
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
df=pd.read_csv('/content/Company_Data (1).csv')
df
```

```python
df.shape[0] # total 400 rows
```

```python
df.shape[1] # total 11 columns
```

```python
df.info() # 0 null values
```

```python
df1=pd.Series(df['Sales'])
s=[]
for i in df1:
    if i>9:
        s.append('good')

    else:
        s.append('bad')

print(s)
```

```python
df2=pd.DataFrame(s)
df3=pd.concat([df2,df],axis=1)
df3=df3.rename({0:'o/p'},axis=1)
df3
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
#create a instance
labelencoder=LabelEncoder()

df3.iloc[:,0]=labelencoder.fit_transform(df3.iloc[:,0])
df3.iloc[:,7]=labelencoder.fit_transform(df3.iloc[:,7])
df3.iloc[:,-1]=labelencoder.fit_transform(df3.iloc[:,-1])
df3.iloc[:,-2]=labelencoder.fit_transform(df3.iloc[:,-2])

df3
```

```python
# split the data into X and Y
array=df3.values

X=array[:,2:10]
Y=array[:,0]
```

```python
num_trees=100
max_features=5
kfold=KFold(n_splits=10,random_state=7,shuffle=True)
model=RandomForestClassifier(n_estimators=num_trees,max_features=max_features)
results=cross_val_score(model,X,Y,cv=kfold)
```

```python
results
```

```python
results.mean()*100
```

**Assignment of Fraud Check**

Use Random Forest to prepare a model on fraud data
treating those who have taxable_income <= 30000 as "Risky" and others are "Good"


```python
# Import Libraries
import pandas as pd
from pandas import read_csv
from sklearn.model_selection import KFold
from sklearn.model_selection  import cross_val_score
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
fraud_check=pd.read_csv('/content/Fraud_check (1).csv')
fraud_check
```

```python
fraud_check.describe()
```

```python
fraud_check.info() # 0 null values
```

```python
a=[]
df=pd.Series(fraud_check['Taxable.Income'])
for i in df:
  if i<=30000:
    a.append('Risky')
  else:
    a.append('Good')

series=pd.DataFrame(a)
series
```

```python
df2=pd.DataFrame(a)
df2=pd.concat([df2,fraud_check],axis=1)
df2=df2.rename({0:'o/p'},axis=1)
df2
```

```python
# Label encoding
from sklearn.preprocessing import LabelEncoder

labelencoder=LabelEncoder()
df2.iloc[:,0]=labelencoder.fit_transform(df2.iloc[:,0])
df2.iloc[:,1]=labelencoder.fit_transform(df2.iloc[:,1])
df2.iloc[:,2]=labelencoder.fit_transform(df2.iloc[:,2])
df2.iloc[:,-1]=labelencoder.fit_transform(df2.iloc[:,-1])
df2
```

```python
#Split the data in X and Y
array=df2.values
X=array[:,1:7]
Y=array[:,0]
```

```python
num_trees=100
max_features=5

kfold=KFold(n_splits=10,random_state=7,shuffle=True)
model=RandomForestClassifier(n_estimators=num_trees,max_features=max_features)
result=cross_val_score(model,X,Y,cv=kfold)
result
```

```python
result.mean()*100
```
